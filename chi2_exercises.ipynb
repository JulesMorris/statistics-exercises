```python
import pandas as pd
from scipy import stats
from pydataset import data
from env import get_db_url
```

```python
contingency_table = pd.DataFrame({'codeup_student':[49,1], 'not_codeup':[20,30]}, index = ['macbook', 'no_macbook'])
contingency_table
```

```python
chi2, p, degf, expected = stats.chi2_contingency(contingency_table)

print('Observed:\n')
print(contingency_table.values)
print('-----------------\nExpected: \n')
print(expected)
print('-------------------\n')
print(f'chi2 = {chi2:.3f}')
print(f'p-value: {p:.3f}') #since p-value is less than alpha, reject null
```

```python
# \alpha + tab
α = 0.05
if p < α:
    print('Reject the null hypotheis.')
else:
    print('Fail to reject the null hypothesis.')
```

## Exercise 2 w/automation

```python
import itertools #will be used for automation the process of categorical vs categorical
```

```python
#are manufacturer and model independent from one another?

df = data('mpg')

#rename class
df = df.rename(columns = {'class': 'vehicle_type'})

#specify automatic as boolean

df["is_automatic"] = df.trans.str.startswith("a")
df.head()
```

```python
#make a list of categorical column names

categorical_columns = ['manufacturer', 'model', 'drv', 'fl', 'vehicle_type', 'is_automatic']
categorical_columns #trans is excluded bc is_automatic answers that, therefore redundant
```

```python
#for combinations, order doesn't matter, for permutations it does

combinations = list(itertools.combinations(categorical_columns, 2))
combinations
print(len(combinations))
```

```python
α = 0.05
outcomes = []

for combination in combinations:
    outcome = {} 
    
    outcome["column1"] = combination[0]
    outcome["column2"] = combination[1]
    
    observed = pd.crosstab(df[combination[0]], df[combination[1]])
    chi2, p, degf, expected = stats.chi2_contingency(observed)
    
    outcome["p"] = p
    if p < α:
        outcome["reject_null"] = True
    else:
        outcome["reject_null"] = False
        
    outcomes.append(outcome)
    
pd.DataFrame(outcomes)
```
